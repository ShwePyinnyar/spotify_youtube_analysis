 Top 10 Artists - with the Highest Views on YouTube

```python
df.sample()
```

```python
df_top_aritst = df.groupby('Artist')['Views'].sum().sort_values(ascending=False).head(10).reset_index(name='Total_View').copy()
df_top_aritst
```

```python
sns.barplot(data=df_top_aritst,x='Artist',y='Total_View')
plt.xticks(rotation=45,ha='right')
plt.ylabel('Total View (billions)')
import matplotlib.ticker as mtick

ax = plt.gca()
ax.yaxis.set_major_formatter(
    mtick.FuncFormatter(lambda x, _: f'{x/1e9:.1f}B')
)
```

These are the top 10 most viewed artists. I grouped them as some of them may have multiple videos.

 Which Top 7 Tracks have the highest Like-to-View ratio on YouTube? 

```python
df.sample()
```

```python
df_ratio = df.groupby('Track').agg({'Likes':'sum','Views':'sum'}).copy()
```

```python
df_ratio.head()
```

```python
df_ratio['Like_View_Ratio'] = df_ratio['Likes']/df_ratio['Views']
df_ratio.head()
```

```python
df_top_ratio = df_ratio.sort_values('Like_View_Ratio',ascending=False).head(7)
df_top_ratio
```

```python
sns.barplot(data=df_top_ratio,x='Track',y='Like_View_Ratio')
plt.xticks(rotation=45,ha='right')
```

What is the Correlation between Views, Likes, Comments, and Stream?

```python
df_corr = df[['Views','Likes','Comments','Stream']].corr()
df_corr
```

```python
sns.heatmap(data=df_corr, cmap='RdYlGn', annot=True)
plt.show()
```

Likes and Views are correlated really well. Other than that, Likes and comments and Likes and Views are the only other ones worth noting. Likes is the common one in correlation between those 3.
